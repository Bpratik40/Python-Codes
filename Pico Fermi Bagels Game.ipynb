```python
import random

def generate_number():
    digit = list(range(10))
    random.shuffle(digit)
    return digit[:3]

def make_guess():
    
    while True:
        guess = input("Enter Guess a 3 Digit Number :")
        if guess.isdigit() and len(guess) == 3:
             return[int(digit) for digit in guess]
        print("Invalid")
        
def check_guess(guess_number, user_guess):
    
    if user_guess == guess_number:
        return "Fermi Fermi Fermi"
        
    hints = []
    for i in range(3):
        if user_guess[i] == guess_number[i]:
             hints.append("Fermi")
        elif user_guess[i] in guess_number:
             hints.append("Pico")
            
    if not hints:
        return "bagles"
        
    return " ".join(hints)
    
def play_game():
    print("Let's Play")
    print("Computer thinking a one number you have to guess it, its a 3 digit number")
    print("Bagels Means NO NUMBER IS CORRECT OR NOT EVEN IN RIGHT POSITION")
    print("Pico Means ONE NUMBER IS CORRECT IN A RIGHT POSITION")
    print("Fermi Means RIGHT NUMBER WITH RIGHT POSITION")
    print("For Example: 234 is right number it will print Fermi Fermi Fermi")
    
    guess_number = generate_number()
    
    while True:
        guess = make_guess()
        output = check_guess(guess_number, guess)
        print(output)
        
        if output == "Right Guess":
            break
            
    print("Game End")
    
play_game()
```
